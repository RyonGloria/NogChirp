```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 80
T = 1/Fs
L = 65
t = np.arange(0, L*T, T)
X = 3*np.cos(2*np.pi*2*t) + 2*np.cos(2*np.pi*4*t) + np.sin(2*np.pi*6*t)
```

```python
Y = np.abs(np.fft.fft(X,len(X)*4))
```

```python
plt.subplot(2,1,1)
plt.plot(np.abs(np.fft.fft(X,len(X))))
plt.subplot(2,1,2)
plt.plot(np.abs(np.fft.fft(X,len(X)*4)))
```

```python
Y = np.fft.fft(X)
f = Fs * np.arange(0, (L-1)//2 + 1) / L
P2 = np.abs(Y / L)
P1 = P2[:((L+1)//2)]
P1[1:] = 2 * P1[1:]

plt.plot(f, P1, "-o")
plt.title("Single-Sided Spectrum of Original Signal")
plt.xlabel("f (Hz)")
plt.ylabel("|P1(f)|")
plt.show()
```

```python
n = len(X) * 4
Y = np.fft.fft(X, n)
f = Fs * np.arange(0, n//2 + 1) / n
P2 = np.abs(Y / L)
P1 = P2[:n//2 + 1]
P1[1:-1] = 2 * P1[1:-1]

plt.plot(f, P1, "-o")
plt.title("Single-Sided Spectrum of Padded Signal")
plt.xlabel("f (Hz)")
plt.ylabel("|P1(f)|")
plt.show()
```
